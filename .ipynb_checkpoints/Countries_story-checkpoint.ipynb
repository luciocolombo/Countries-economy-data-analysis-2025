```python
sns.boxplot(data=df,x='region',y='GDP', showfliers=False)
print('The following boxplot shows that the median Asia country has a GDP similar to America, but both Asia and Europe have higher GDP countries (represented by the box over the line in each box) ')
```

```python
#Do landlocked countries (borders present, no coastline) have systematically different GDP per capita than coastal countries?
sns.boxplot(x=df['borders'].notna().map({True:'Land', False:'Island'}), y=df['GDP'], showfliers=False)
print('This questions cannot be answered with the current information. We do not have information about having or not having coasts. We can only compare islands with no islands')
print(df['borders'].notna().sum(), 'Land, ', df['borders'].isna().sum(), 'Island')
print('Yes, Land countries have significantly higher median and higher GDP countries overall.')
df.groupby(df.borders.notna().map({True:'Land',False:'Island'})).GDP.median().reset_index().rename(columns={'GDP':'Median GDP'})
```

```python
#How does inflation rate differ between high-income vs low-income regions?
sns.lineplot(data=df, x='GDP', y='Inflation Rate')
print('There is no clear mathematical relationship. What is clear is that only poor countries (GDP<2500) have very high inflations (more than 5%). Inflation is correlated with poverty. We cannot imply causation')
```

```python
# Mask for poor countries
poor_countries = df[df['GDP'] < 2500]

# Count how many have high vs low inflation
n_poor_high_inflation = poor_countries[poor_countries['Inflation Rate'] > 5].shape[0]
n_poor_low_inflation  = poor_countries[poor_countries['Inflation Rate'] <= 5].shape[0]

# Calculate percentage of high inflation among poor countries
rate_poor_high_inflation = round(100 * n_poor_high_inflation / poor_countries.shape[0], 1)

# Output
print(f"There are {n_poor_high_inflation} countries with GDP < 2500 and inflation > 5%. That is {rate_poor_high_inflation}% of poor countries.")
print(f"There are {n_poor_low_inflation} countries with GDP < 2500 and inflation ‚â§ 5%.")
```

```python
#Which region has the highest average debt-to-GDP ratio?
sns.barplot(data=df, x='region', y='Debt/GDP')


max_debt = df.groupby(['region'])['Debt/GDP'].mean().idxmax()
print(f'All regions seem to have a very similar Debt/GDP ratio. Difference do not look significant. Highest is {max_debt}')
df.groupby(['region'])['Debt/GDP'].mean()
```

```python
#Do countries with higher population tend to have higher GDP growth?
ds = df[df['Population']<1200] # We remove 2 extreme outliers that will skew the sample results.
sns.regplot(data=ds,x='Population',y='GDP Growth')

print('There is a very weak positive correlation between population and GDP growth. If we compute the correlation index, we get 0.27. Barely related')
print('Note: 42% of GDP Growth values are missing, so this result is not completely conclusive.')
ds[['Population','GDP Growth']].corr()
```

```python
#Identify outliers: countries with unusually high debt/GDP compared to their peers.
df[['Debt/GDP','name']].sort_values('Debt/GDP', ascending=False).head()
```

```python
#¬øLa deuda afecta el crecimiento?
sns.regplot(x=df['Debt/GDP'],y=df['GDP Growth'])
df[['Debt/GDP','GDP Growth']].corr()
print('There is a weak correlation between debt and less GDP growth. The correlation coeficient is -0.29, considered weak.')
print('Note: 42% of GDP Growth values are missing, so this result is not completely conclusive.')
```

```python
#While not strictly necessary, we can confirm the result by modeling this behavior with a linear regression as well and calculate the R^2 error directly
ds = df.dropna()
Y = ds['GDP Growth']
X = ds[['Debt/GDP']]
model = LinearRegression().fit(X,Y)


y = model.predict(X)

sns.lineplot(x=X['Debt/GDP'],y=y)
sns.scatterplot(x=ds['Debt/GDP'], y=ds['GDP Growth'], color="blue")

print(model.score(X,Y),
model.coef_,
model.intercept_)

print('The linear regression has an error score of 0.086, meaning it explains only 8,5% of the variance. The correlation is very weak.')
print('Note: 42% of GDP Growth values are missing, so this result is not completely conclusive.')
```

```python
#Demostrar que no hay relacion entre GDP y LATITUD
#We will model the problem with a linear regression to prove they are not related.
ds = df[df['GDP']<10000] #filter out outliers 
X = ds[['GDP']].dropna()
Y = ds['latitude'].dropna()
model3 = LinearRegression().fit(X,Y)
y = model3.predict(X)
sns.lineplot(x=X['GDP'],y=y)
sns.scatterplot(x=X['GDP'],y=Y)
print('There is no correlation, as seen with the score obtained:', round(model3.score(X,Y)*100,1), '%')

```

```python
#If you calculate GDP per capita (GDP / Population), which countries look strongest/weakest relative to raw GDP?
df['GDP_capita'] = df['GDP'] / df['Population']
df['Gdp_VS_GDPCapita_Ranking'] = 1/df['Population'] #Because it is GDP/(GDP/population) = 1/population

print('We see that the least unbalanced contries are China, India and US, and the most unbalanced are Monaco, Liechtenstein, Cayman Islands, Seychelles')
print('There is a pattern where very high populations rank very well, and small islands do not. This could be studied further')
df[['Gdp_VS_GDPCapita_Ranking','name', 'GDP', 'GDP_capita']].sort_values('Gdp_VS_GDPCapita_Ranking').rename(columns={'Gdp_VS_GDPCapita_Ranking':'Inbalance GDP vs GDP per capita index', 'name':'Country', 'GDP_capita':'GDP per capita'})
```

## üß† 5. Clustering Analysis (K-means)
**Objective:** group countries with similar economic profiles.


```python
#Cluster countries by GDP, inflation, and jobless rate ‚Äî what groups emerge?
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
km_model = KMeans(n_clusters=4)
X = df[['GDP','Jobless Rate','Inflation Rate']].dropna()
Xscaled = scaler.fit_transform(X)

km_model = km_model.fit(Xscaled)
```

```python
## The following graph shows the elbow of the inertia function, used to search for the best K value for the KMeans clustering method. It's not a study on the dataset.
inertia = []
kamount = []
for k in range(1,10):
    temp_kmeans = KMeans(n_clusters=k).fit(X)
    inertia.append(temp_kmeans.inertia_)
    kamount.append(k)

sns.lineplot(x=kamount,y=inertia)
plt.xlabel('k')
plt.ylabel('inertia')
print('We decide to choose a value of K=4, meaning 4 categories in our data.')
```

```python
df1 = df[['GDP','Jobless Rate','Inflation Rate', 'name']].dropna()
df1['cluster'] = km_model.labels_
#df1[df1['cluster']==0]
df1['cluster'].value_counts()
#We see most countries in cluster 0. Then there are 3 rare categories with exceptions
```

```python
plt.figure(figsize=(15,7))
sns.scatterplot(data=X, x='GDP', y='Jobless Rate', hue=km_model.labels_, palette="Set2")
print('For ilustration purposes, we will show some graphs on 2D to better visualize the last 3D interactive graph.')
```

```python
print('After the in depth study of the clusters, we can rename the clusters with better names. This is done now so the following is easier to understand')
df1.cluster = df1.cluster.map({2:'GDP Titans', 1: 'Inflationary', 3:'Low development', 0:'Average country' })
```
